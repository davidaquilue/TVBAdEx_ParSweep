# Understanding K means

A simplified approach to the classification of FC or corrFCSC traces to have a better understanding of the logic behind the processes.

```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
import os

from JUSUFlike.Project.Codes.processing_results import *
from IPython.display import Image
import tvb_model_reference.src.nuu_tools_simulation_human as tools
from sklearn.cluster import KMeans

SMALL_SIZE = 16
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Computer Modern Roman"]})

fig_labels = ['$S$', '$b_e$ (pA)', '$E_{L,i}$ (mV)', '$E_{L,e}$ (mV)', '$T$ (ms)']
```

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 16
BIGGER_SIZE = 18
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
FC_mat = np.array(([[0.8, 0.45, 0.2, 0], [0.75, 0.3, 0.25, 1], 
                    [0.13, 0.8, 0.93, 2], [0.24, 0.7, 0.68, 3]]))  # FC at b=0, FC at b=120, label for title
legend = ['$S$ = 0.3, $E_{L,i}$ = -80 mV, $E_{L,e}$ = -76 mV, $T$ = 19 ms',
         '$S$ = 0.35, $E_{L,i}$ = -76 mV, $E_{L,e}$ = -76 mV, $T$ = 26 ms',
         '$S$ = 0.25, $E_{L,i}$ = -60 mV, $E_{L,e}$ = -60 mV, $T$ = 19 ms',
         '$S$ = 0.4, $E_{L,i}$ = -64 mV, $E_{L,e}$ = -60 mV, $T$ = 26 ms']

# First the FC(b) traces
fig, ax = plt.subplots(1, 1)
for i in range(FC_mat.shape[0]):
    ax.plot([0, 60, 120], FC_mat[i, :3], label=legend[i])
ax.set(ylim=(0, 1), xlabel='$b_e$', ylabel='$FC(b_e)$', title='$FC(b_e)$ traces for 4 parametrizations')
#ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
#          fancybox=True, shadow=True, ncol=1)
ax.legend(loc=(1.05, 0.25))
fig.savefig('./figures/figures_report/kmeans/traces.pdf', bbox_inches='tight')
plt.show()
```

```python
#fig, ax = plt.subplots(1, 1)
from matplotlib.ticker import MaxNLocator

SMALL_SIZE = 20
MEDIUM_SIZE = 22
BIGGER_SIZE = 24
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')

# Perform the K-means:
kmeans_FC = KMeans(n_clusters=2, random_state=4).fit(FC_mat[:, :3])
centers = kmeans_FC.cluster_centers_

for i in range(FC_mat.shape[0]):
    ax.scatter(FC_mat[i, 0], FC_mat[i, 1], FC_mat[i, 2], label=legend[i], s=42)
    
ax.scatter(centers[0, 0], centers[0, 1], centers[0, 2], label='Class 0 centroid', c='y', marker='*', s=42)
ax.scatter(centers[1, 0], centers[1, 1], centers[1, 2], label='Class 1 centroid', c='c', marker='*', s=42)
ax.set(ylim=(-0.1, 1.1), xlim=(-0.1, 1.1), xlabel='$FC(b_e = 0 pA)$', ylabel='$FC(b_e = 60 pA)$', 
       title='K-means space', zlabel='$FC(b_e = 120 pA)$', zlim=(-0.1, 1))
#ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
#          fancybox=True, shadow=True, ncol=1)
ax.legend(loc=(1.3, 0.25))
ax.set_xlabel('$FC(b_e = 0 pA)$', labelpad=10)
ax.set_ylabel('$FC(b_e = 60 pA)$', labelpad=10)
ax.set_zlabel('$FC(b_e = 120 pA)$', labelpad=10)
ax.xaxis.set_major_locator(MaxNLocator(4)) 
ax.yaxis.set_major_locator(MaxNLocator(4)) 
ax.zaxis.set_major_locator(MaxNLocator(4)) 
fig.savefig('./figures/figures_report/kmeans/kspace.pdf',  bbox_inches='tight')
plt.show()
```

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 16
BIGGER_SIZE = 18
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
fig, ax = plt.subplots(1, 1)
for i in range(FC_mat.shape[0]):
    ax.plot([0, 60, 120], FC_mat[i, :3], label=legend[i])
ax.plot([0, 60, 120], centers[0, :], label='Class 0 centroid', c='y')
ax.plot([0, 60, 120], centers[1, :], label='Class 1 centroid', c='c')
ax.set(ylim=(0, 1), xlabel='$b_e$', ylabel='$FC(b_e)$', title='$FC(b_e)$ traces for 4 parametrizations and centoids')
#ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
#          fancybox=True, shadow=True, ncol=1)
ax.legend(loc=(1.05, 0.25))
fig.savefig('./figures/figures_report/kmeans/traces_centroids.pdf',  bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
for i in range(FC_mat.shape[0]):
    if i < 2:
        ax.plot([0, 60, 120], FC_mat[i, :3], label=legend[i], c='c')
    else:
        ax.plot([0, 60, 120], FC_mat[i, :3], label=legend[i], c='y')
ax.set(ylim=(0, 1), xlabel='$b_e$', ylabel='$FC(b_e)$', title='Classified $FC(b_e)$ traces for 4 parametrizations')
#ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),
#          fancybox=True, shadow=True, ncol=1)
ax.legend(loc=(1.05, 0.25))
fig.savefig('./figures/figures_report/kmeans/traces_classified.pdf',  bbox_inches='tight')
plt.show()
```
